```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
air_bnb = pd.read_csv('../Friday_Project/AB_NYC_2019.csv')
air_bnb.head()
```

```python
# How many neighborhood groups are available and which shows up the most?

air_bnbgroups= air_bnb['neighbourhood_group'].describe()
print(air_bnbgroups)
print(f'These are the neighborhood groups available {air_bnb.neighbourhood_group.unique()}')
print('Manhattan shows up the most')
```

```python
# Are private rooms the most popular in manhattan?

rooms =air_bnb.groupby(by=['neighbourhood_group','room_type']).count()

print(f'No, Private Rooms are most popular in Brooklyn ')
print(rooms)
```

```python
# Which hosts are the busiest and based on their reviews?
busiest=air_bnb[['host_name','number_of_reviews']]
groupedhost = busiest.groupby('host_name')['number_of_reviews'].count()

groupedhost.sort_values(ascending=False)
```

```python
#Which neighorhood group has the highest average price?
avgprivce= air_bnb[[ 'neighbourhood_group','price']]
avgprivce.groupby(by='neighbourhood_group').mean()
```

```python
# Which neighbor hood group has the highest total price?
avgprivce.groupby(by='neighbourhood_group').sum()
```

```python
#Which top 5 hosts have the highest total price?
totalloc= air_bnb
# totalloc.groupby(['host_name'],as_index=False).sum().sort_values('price',ascending=False)
grouped=totalloc.groupby('host_name')['price'].sum()
grouped.sort_values(ascending=False).head(6)
```

```python
# Who currently has no (zero) availability with a review count of 100 or more?
air_bnb['number_of_reviews']>=100
morethan= air_bnb[air_bnb['number_of_reviews']>=100]
morethan.sort_values(['number_of_reviews'],ascending=False).drop_duplicates()

```

```python
# What host has the highest total of prices and where are they located?
grouped=totalloc.groupby(by=['host_name','neighbourhood_group'])['price'].sum()
grouped.sort_values(ascending=False).head(1)
# Sonder  has the highest total of prices
```

```python
# When did Danielle from Queens last receive a review?

doesdan = air_bnb[air_bnb['host_name']=='Danielle']
doesdan.sort_values('last_review',ascending=False).head(10)
# She rlast received one on 2019-07-08	
```
